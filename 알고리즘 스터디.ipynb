```python
#치킨배달

from itertools import combinations

n,m=map(int,input().split()) #도시크기n, 남길 치킨집 수m
house=[] #집
chi=[]   #치킨집
for i in range(n):#행을 돌면서
    for j,val in enumerate(map(int,input().split())): #인덱스(열)와 값을 추출
        if val==1: #집이라면
            house.append((i,j)) #집좌표 저장
            
        elif val==2: #치킨집이라면
            chi.append((i,j))  #치킨집 좌표 저장
            
def get_dist(h,c): #거리 구하기
    r1,c1=h 
    r2,c2=c
    return abs(r1-r2)+abs(c1-c2) #집과 치킨집의 맨해튼 거리

ans=2*n*len(house) #최대거리값으로 초기화
for comb in combinations(chi,m): #남는 치킨집의 조합으로 구함
    tot=0
    for h in house: #집을 돌면서
        tot+=min(get_dist(h,chi) for chi in comb) #집과 치킨집의 거리 구함
    ans=min(ans,tot)
print(ans) #치킨집을 m개 남겼을 때의 조합을 모두 계산한다음 최소값 반환
```
